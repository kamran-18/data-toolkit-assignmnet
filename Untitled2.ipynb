```python
#questionThe difference in npYarray, npYasarray and npYasanyarray
#answer:
#np.array()
'''This function creates a new NumPy array from an existing Python list or tuple.
If the input is already a NumPy array, it makes a copy of that array.
It always returns a new array.
The data type of the resulting array is determined by the data type of the input'''

#np.asarray()
'''This function is similar to np.array(), but it tries to reuse the input array if possible rather than making a copy.
If the input is already a NumPy array, it does not create a new array but returns the input array itself.
If the input is a Python list or tuple, it creates a new NumPy array.'''


#np.asanyarray()
'''This function is similar to np.asarray(), but it also accepts subclasses of ndarray.
It returns an ndarray if the input is a subclass of ndarray, otherwise, it returns a new ndarray.'''
```

```python
#The difference between Deep copy and shallow copy

#let me explain this by means of an example
#deepcopy
l=[1,2,3,4]
arr=np.array(l)
print(arr)
#now le me make a copy of above
a=arr
print(a)
#now i will change arr i ll add some thing in it
arr[0]=120
print(arr)
print(a)
# if we see here the change in arr reflects the change is a this is known as deep copy
# here there is same memory allocation

#swallow copy

b=a.copy()
b[0]=2000
print('output of swallow copy:',b)

print(a)


# here if we the change in b dosnt reflect in the change in a
#it is called swallow copy as the memory allocation here is different
```

```python
#question 14
#1
# Generate random data for the time-series
dates = pd.date_range('2024-05-20', periods=1000)
temperature = np.random.normal(loc=20, scale=5, size=1000)
humidity = np.random.normal(loc=50, scale=10, size=1000)

# Create the DataFrame
df = pd.DataFrame({'Date': dates, 'Temperature': temperature, 'Humidity': humidity})
```

```python
fig, ax1 = plt.subplots()

# Plot temperature on the left y-axis
color = 'tab:red'
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature', color=color)
ax1.plot(df['Date'], df['Temperature'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

# Create a second y-axis for humidity
ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('Humidity', color=color)
ax2.plot(df['Date'], df['Humidity'], color=color)
ax2.tick_params(axis='y', labelcolor=color)

plt.title('Temperature and Humidity Over Time')
fig.tight_layout()
plt.show()
```

```python
#question 15
import numpy as np
import matplotlib.pyplot as plt

# Create a NumPy array 'data' containing 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)

# Plot a histogram of the data with 30 bins
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Histogram')

# Calculate the mean and standard deviation of the data
mean = np.mean(data)
std_dev = np.std(data)

# Generate values for the PDF
x = np.linspace(-4, 4, 1000)
pdf = (1/(std_dev * np.sqrt(2 * np.pi))) * np.exp(-(x - mean)**2 / (2 * std_dev**2))

# Overlay a line plot representing the normal distribution's probability density function (PDF)
plt.plot(x, pdf, color='r', linestyle='-', linewidth=2, label='PDF')

# Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

# Set the title of the plot as 'Histogram with PDF Overlay'
plt.title('Histogram with PDF Overlay')

# Display the legend
plt.legend()

# Show the plot
plt.show()
```

```python
# Create two random arrays
x = np.random.randn(100)
y = np.random.randn(100)

# Determine the quadrant for each point
quadrant = np.where((x >= 0) & (y >= 0), 'Quadrant I',
                    np.where((x < 0) & (y >= 0), 'Quadrant II',
                             np.where((x < 0) & (y < 0), 'Quadrant III', 'Quadrant IV')))

# Create a DataFrame for the scatter plot
df_scatter = pd.DataFrame({'x': x, 'y': y, 'Quadrant': quadrant})

# Set up the color palette for each quadrant
palette = {'Quadrant I': 'blue', 'Quadrant II': 'green',
           'Quadrant III': 'orange', 'Quadrant IV': 'red'}

# Create the scatter plot
sns.scatterplot(x='x', y='y', hue='Quadrant', palette=palette, data=df_scatter)
plt.legend(title='Quadrants')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')
plt.show()
```

```python
import numpy as np
from bokeh.plotting import figure, show

# Generate data for the sine wave function
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a new plot
p = figure(title='Sine Wave Function', x_axis_label='X-axis', y_axis_label='Y-axis')

# Plot the sine wave function
p.line(x, y, line_width=2)

# Add grid lines
p.grid.grid_line_color = "black"
p.grid.grid_line_alpha = 0.5

# Show the plot
show(p)
```

```python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, HoverTool


# Generate random categorical data and corresponding values
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(1, 10, size=len(categories))

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(categories=categories, values=values, color=['blue', 'green', 'orange', 'red', 'purple']))

# Create a new plot
p = figure(x_range=categories, title='Random Categorical Bar Chart', x_axis_label='Categories', y_axis_label='Values')

# Plot the bars
p.vbar(x='categories', top='values', color='color', width=0.5, source=source)

# Add hover tooltips to display exact values
hover = HoverTool(tooltips=[('Category', '@categories'), ('Value', '@values')])
p.add_tools(hover)

# Set the axis labels
p.xaxis.major_label_orientation = np.pi / 4

# Show the plot
show(p)
```

```python
import plotly.graph_objs as go


# Generate random data for the line plot
x = np.linspace(0, 10, 100)
y = np.random.randn(100)

# Create a Plotly trace for the line plot
trace = go.Scatter(x=x, y=y, mode='lines')

# Create the layout for the plot
layout = go.Layout(title='Simple Line Plot', xaxis=dict(title='X-axis'), yaxis=dict(title='Y-axis'))

# Create the figure object
fig = go.Figure(data=[trace], layout=layout)

# Display the plot
fig.show()
```

```python
import plotly.graph_objs as go

# Generate random data for the pie chart
labels = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(1, 10, size=len(labels))

# Create the pie chart trace
trace = go.Pie(labels=labels, values=values, hoverinfo='label+percent', textinfo='value+percent')

# Create the layout for the pie chart
layout = go.Layout(title='Interactive Pie Chart')

# Create the figure object
fig = go.Figure(data=[trace], layout=layout)

# Display the plot
fig.show()
```
